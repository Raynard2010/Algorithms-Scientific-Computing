```python
import numpy as np

Xs = [
    [ [4, 2, -1],
      [1, -5, 2],
      [2, -1, -4]
    ],
    [ [3, 4, 5],
      [-3, 7, -4],
      [1, -4, -2]
    ],
    [ [9, -2, 3, 2],
      [2, 8, -2, 3],
      [-3, 2, 11, -4],
      [-2, 3, 2, 10] ]
     ]

Ys = [
    [41, -10, 1],
    [34, -32, 62],
    [55, -14, 12, -21]
     ]

def diagonally_dominant(x):
  x = np.array(x)

  diag = np.abs(np.diag(x))

  off_diag = np.sum(np.abs(x), axis = 1) - diag

  if(np.all(diag > off_diag)):
    return True
  else:
    return False

def gauss_seidel(x, y, max = 15, tolerance = 0.022):
  x = np.array(x)
  y = np.array(y)

  n_old = np.zeros(len(x[0]))

  diag = np.diag(x)

  x = -x

  np.fill_diagonal(x, 0)

  for iterasi in range(max):
    n_new = np.array(n_old)

    for i in range(len(x)):
      n_new[i] = (y[i] + np.dot(x[i], n_new)) / diag[i]

    print('Iterasi', iterasi + 1, n_new)

    euclidean = np.sqrt(np.dot(n_new - n_old, n_new - n_old))

    if(euclidean < tolerance):
      return True

    n_old = n_new

  return False

for i,(X, Y) in enumerate(zip(Xs, Ys)):
  print(f'A: {X} y = {Y}')

  if(diagonally_dominant(X)):
    print('Diagonally Dominant')

    if(gauss_seidel(X, Y)):
      print('Konvergen')
    else:
      print('Not konvergen')
  else:
    print('Not diagonally dominant')

  print('\n')
```
